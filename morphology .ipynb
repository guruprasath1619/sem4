# MORPHOLOGY TRANSFORMATION AND IMAGE SEGMENTATION

# DILATION

```python
import cv2 
import numpy as np
```

```python
from matplotlib import  pyplot as plt
```

```python
img=cv2.imread("C:/Users/admin/Downloads/ney.jpg")
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
_,mask = cv2.threshold(img,220,255,cv2.THRESH_BINARY)
kernel=np.ones((3,3),np.uint8)
dilation=cv2.dilate(mask,kernel,iterations=1)
erode=cv2.erode(mask,kernel,iterations=1)
titles=['image','mask','dilate','erode']
images=[img,mask,dilation,erode]
for i in range(len(images)):
    plt.subplot(2,2,i+1),plt.imshow(images[i])
    plt.title(titles[i])
plt.imshow
```

# MRPHOLOGICAL OPENINGH

```python
open=cv2.morphologyEx(img,cv2.MORPH_OPEN,kernel)
```

```python
cv2.imshow("image",img)
cv2.imshow('open',open)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
close=cv2.morphologyEx(img,cv2.MORPH_CLOSE,kernel)
```

```python
cv2.imshow("image",img)
cv2.imshow('close',close)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# morphological gradient-erosion,dilation and difference both

```python
gradient=cv2.morphologyEx(img,cv2.MORPH_GRADIENT,kernel)
```

```python
cv2.imshow("image",img)
cv2.imshow('morph gradient',gradient)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
blackhat=cv2.morphologyEx(img,cv2.MORPH_BLACKHAT,kernel)
```

```python
cv2.imshow("image",img)
cv2.imshow('blackhat',blackhat)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
tophat=cv2.morphologyEx(img,cv2.MORPH_TOPHAT,kernel)
```

```python
cv2.imshow("image",img)
cv2.imshow('tophat',tophat)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
